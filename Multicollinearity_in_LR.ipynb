### Multicollinearity in Linear Regression

```python
### Get the data 

url ='https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/Advertising.csv'
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv(url, index_col=0)
```

```python
df.head()
```

```python
X = df.iloc[:, :-1]
```

```python
y = df.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
X = sm.add_constant(X)
```

```python
### fit a OLS model with intercept on TV and Radio

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:, 1:].corr()
```

```python
salary_data = 'https://raw.githubusercontent.com/krishnaik06/Multicollinearity/master/data/Salary_Data.csv'
```

```python
df_salary = pd.read_csv(salary_data)
```

```python
df_salary.head()
```

```python
X = df_salary.iloc[:, :-1]
y = df_salary.iloc[:, -1]
```

```python
X.head()
```

```python
y.head()
```

```python
### fit a OLS model with intercept on TV and Radio

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
model.summary()
```

#### Very High std error therefore we can say there is definately multicollinearity.

```python
X.iloc[:, 1:].corr()
```
